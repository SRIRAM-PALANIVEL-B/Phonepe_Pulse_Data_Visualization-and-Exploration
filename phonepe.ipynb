```python
import os
import json
import pandas as pd
import psycopg2
```

```python
# Aggregated_Transaction


path1 = "D:/SRIRAM/PROJECTS (CAPSTONE)/PROJECTS/MINI/Capstone Project/PHONEPE/pulse/data/aggregated/transaction/country/india/state/"
aggre_trans_list = os.listdir(path1)

column1 = {"States":[], "Years":[], "Quarter":[], "Transaction_Type":[], "Transaction_Count":[], "Transaction_Amount":[]} 


for state in aggre_trans_list:
    current_state = path1 + state + "/"
    aggre_year_list = os.listdir(current_state)
    

    for year in aggre_year_list:
        current_year = current_state + year +"/"
        aggre_file_list = os.listdir(current_year)
        

        for file in aggre_file_list:
            current_file = current_year + file
            data = open(current_file,"r")


            A = json.load(data)

            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                column1["Transaction_Type"].append(name)
                column1["Transaction_Count"].append(count)
                column1["Transaction_Amount"].append(amount)
                column1["States"].append(state)
                column1["Years"].append(year)
                column1["Quarter"].append(int(file.strip(".json")))
```

```python
# change details of dict into dataframe

aggree_transaction = pd.DataFrame(column1)
```

```python
#Change Name and Detais in DataFrame (Aligned for GeoCode)
aggree_transaction["States"] = aggree_transaction["States"].str.replace("andaman-&-nicobar-islands", 'Andaman & Nicobar islands')
aggree_transaction["States"] = aggree_transaction["States"].str.replace("-", " ")
aggree_transaction["States"] = aggree_transaction["States"].str.title()
aggree_transaction["States"] = aggree_transaction["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and nagar Haveli and Daman and Diu')
```

```python
# Aggregated_User


path2 = "D:/SRIRAM/PROJECTS (CAPSTONE)/PROJECTS/MINI/Capstone Project/PHONEPE/pulse/data/aggregated/user/country/india/state/"
aggre_user_list = os.listdir(path2)


column2 = {"States":[], "Years":[], "Quarter":[], "Brands":[], "Transaction_Count":[], "Percentage":[]} 


for state in aggre_user_list:
    current_state = path2 + state + "/"
    aggre_year_list = os.listdir(current_state)
    

    for year in aggre_year_list:
        current_year = current_state + year +"/"
        aggre_file_list = os.listdir(current_year)
        

        for file in aggre_file_list:
            current_file = current_year + file
            data = open(current_file,"r")


            B = json.load(data)
            
            try:
                for i in B["data"]['usersByDevice']:
                    brand = i["brand"]
                    count = i['count']
                    percentage = i['percentage']
                    column2["Brands"].append(brand)
                    column2["Transaction_Count"].append(count)
                    column2["Percentage"].append(percentage)
                    column2["States"].append(state)
                    column2["Years"].append(year)
                    column2["Quarter"].append(int(file.strip(".json")))

            except:
                pass
```

```python
# Connect SQL

mydb = psycopg2.connect(host = "localhost",
                        user = "postgres",
                        port = "5432",
                        database = "phonepe_capstone",
                        password = "12345678")
cursor = mydb.cursor()
```
